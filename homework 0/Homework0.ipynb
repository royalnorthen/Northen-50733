```python
# code

from matplotlib import pyplot as plt
import numpy as np

#raw data
data = np.genfromtxt("fakeData.txt", delimiter=",")
x = data[:,0]
y = data[:,1]
bad = data[:,2]

#filtered data 
x_filtered = x[(bad==1)]
y_filtered = y[(bad==1)]

x_bad = x[(bad!=1)]
y_bad = y[(bad!=1)]

#plotting
fig1 = plt.figure(figsize=(20,10))
ax1 = fig1.add_subplot(111)
ax1.scatter(x_filtered,y_filtered,s=15,c='b',alpha=0.5)
ax1.scatter(x_bad,y_bad,s=50,c='r',marker='x',alpha=1.0)

ax1.xaxis.set_tick_params(labelsize=20)
ax1.yaxis.set_tick_params(labelsize=20)

ax1.set_xlim(0,120)
ax1.set_ylim(0,12000)
```
